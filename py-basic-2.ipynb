11. In a for loop, what is the difference between range(10), range(0, 10), and range(0, 10, 1)?

Range(10)  will start from 0 and go upto 9. Here starting point is taken as 0 by default.
Range(0,10) is same as range(10) but here starting point is defined by the programmer. 
Range(0,10,1) will also give the same output but 1 specifies the gap of numbers from 0-10.
